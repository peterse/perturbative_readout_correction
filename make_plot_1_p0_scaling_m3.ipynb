### Perform M3 analysis with pre-existing package

We will use the `mthree` package to automate the truncation and correction associated with an experimental response matrix. The basic idea behind M3 is to restrict the response matrix to the subspace that the _estimated_ probability vector has support on. For example, let $p_{true}$ be the true probability vector where $(p_{true})_i$ is the probability of observing each outcome bitstring $i$. As in the paper, $p' := Rp_{true}$ is the distribution describing the bitflip-error measurement distribution. To define an empirical probability distribution, we will sample values in $\{0,1\}^n$ according to $p'$. Letting $n_i$ be the number of times string $i \in \{0,1\}^n$ is obsreved, we define
$$
\hat{p}_i := n_i / N.
$$


Then we compare the readout error mitigation methods by checking their ability to recover $p_{true}$ from $\hat{p}_i$.

```python
# For the distribution, I'm going to use a normal distribution centered at 0000 in the permuted space
# i.e. treat x-values as int_n's that overflow 2^n - 1 -> 0

n = 5
def gaussian(x, x0, sigma):
    return np.exp(-(x - x0)**2 / (2 * sigma**2)) / (np.sqrt(2 * np.pi) * sigma)

def uniform(n):
    return np.ones(1 << n) / (1 << n)

def permuted_nbit_guassian(j, n, sigma):
    """Compute the gaussian centered at n//2 on a bit value `j` that has been shifted by n//2.
    
    This has the effect of centering the gaussian on `0` but maintaining tails symmetrically around
    `000` in both the high- and low-weight subspaces.
    """
    halfpoint = 1 << (n - 1)
    fullrange = 1 << n
    x = ((j + (halfpoint)) % fullrange) / fullrange
    x0 = halfpoint / fullrange
    out = gaussian(x, x0, sigma)
    return out / sum(out)
    
    
def nbit_guassian(j, n, sigma):
    """Compute the asymmetric gaussian centered at 0.
    """
    fullrange = 1 << n
    x = j / fullrange
    x0 = 0
    out = gaussian(x, x0, sigma)
    return out / sum(out)
    
# permuted_nbit_guassian(0, 4, .25)

# gaussian(0, 0, .25)
```

## Construct test R

We enforce that $R$ has a tensor structure so that multi-flip events are necessarily exponentially suppressed in weight change of bitstring.

```python
import gzip
import lzma
import pickle

def load_compressed(filepath, mode='gzip', serial_mode='pickle'):
    """
    Loads a compressed object

    mode: 'gzip' (default), 'lzma', 'raw'
    """
    if serial_mode == 'dill':
        if imported_dill:
            load_f = dill.load
        else:
            RuntimeError('`dill` not found.')
    else:
        load_f = pickle.load

    if mode == 'gzip':
        f = gzip.open(filepath, 'rb')
    elif mode == 'lzma':
        f = lzma.open(filepath, 'rb')
    else:
        f = open(filepath, 'rb')
    obj = load_f(f)
    f.close()
    return obj
```

```python
wvals = [1, 2, 3, 4]
# nqubits_lst = [6, 8, 10, 12]
nqubits_lst = [12]
# R = response_matrix.Rmatrix(q01_arr, q10_arr)
R0 = load_compressed('ibm_cairo_confusion_matrix_12q_10000s_value.lz', mode='lzma')
R1 = load_compressed('ibm_hanoi_confusion_matrix_12q_10000s_value.lz', mode='lzma')
R2 = load_compressed('ibmq_toronto_confusion_matrix_12q_10000s_value.lz', mode='lzma')
R_list = [R0, R1, R2]
# Rinv = np.linalg.inv(R)
```

## Compute two different corrections

```python
# comparison = lambda x,y: scipy.stats.entropy(x,y)
def rel_err(x, y): return abs(x - y) / x # x is 'truth'
def abs_err(x, y): return abs(x - y)
comparison = rel_err

nexperiments = 3
n_qubits = 12


# Outcomes for correcting p0
p0_baseline = np.zeros((nexperiments, len(R_list)))
p0_outcomes = np.zeros((nexperiments, len(wvals), len(R_list))) # rows correspond to w=1,2...
# qfixed = 0.06

for i in range(nexperiments):
    for j, R in enumerate(R_list):
        # Generate response matrix and artificial readout error
#         R = response_matrix.generate_characteristic_R(qfixed, n_qubits)
        x = np.arange(1 << n_qubits)
        if i == 0:
            p_true = permuted_nbit_guassian(x, n_qubits, .25)
        elif i == 1:
            p_true = nbit_guassian(x, n_qubits, .25)
        elif i == 2:
            p_true = uniform(n_qubits)

    #     p_true = nbit_guassian(x, n_qubits, .25)
        p_prime = R @ p_true

        # Just p0 correction
        p0_baseline[i, j] = comparison(p_true[0], p_prime[0])
        for k, w in enumerate(wvals):
            p0_corr = response_matrix.invert_p0_truncated(R, p_prime, w=w)
            p0_outcomes[i, k, j] = abs(p_true[0] - p0_corr)
```
